##Outliers Detection

```python
data = [10, 20, 30, 40, 50, 60, 70, 80 , 90, 1000]

# IQR = Inter Quartile Range

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1
# print(IQR)

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

print("Lower limit = " , lower)
print("Upper limit = " , upper)
```

```python
for d in data:
  if d<lower or d>upper:
    print(d)
```
