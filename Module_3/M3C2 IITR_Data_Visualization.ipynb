```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    "month": ["Jan","Feb","Mar","Apr","May"],
    "sales": [100,150,130,180,200],
    "profit": [20,30,25,40,50],
    "category": ["A","B","A","B","A"]
})
```

```python
plt.plot(df["sales"])

plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
plt.plot(df["month"], df["sales"])
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
plt.plot(df["month"], df["sales"])
plt.plot(df["month"], df["profit"])
plt.show()
```

```python
plt.plot(df["month"], df["sales"])
plt.plot(df["month"], df["profit"])
plt.grid(True)
plt.show()
```

```python
plt.plot(df["month"], df["sales"], color = 'red', label = 'Sales' )
plt.plot(df["month"], df["profit"], color = 'green', label = 'Profit' )
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5 point summary

# max
# 75th percentile
# 50th percentile (median)
# 25th percentile
# min
```

```python
sales = np.array([20, 30, 40, 50, 40, 80 , 1000, 90, 10])

Q1 = np.percentile(sales, 25)
Q3 = np.percentile(sales, 75)

IQR = Q3-Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

print(lower)
print(upper)
plt.boxplot(sales)
plt.show()
```
