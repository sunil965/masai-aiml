#ML Pipeline

```python
1. Raw data
2. Cleaning & Pre-processing --> most important step
3. Feature Engineering
4. Training
5. Evaluation
6. Deployment
```

#What is the purpose of each of the step?

```python
1.Data collection --> What information do we have?
2. Exploring the data --> What does the data look like?
3. Clean data --> is the data ready to be used?
4. Feature Engineering --> Can we further enhance the quality of the data?
5. Training the model --> learning the patterns present in the data.
6. Evaluation --> is the model relaible?
7. Deployment --> making it useful for the target audience.
8. Monitor --> Is everything on track? Any updates needed?
```

#Importance of Every step

```python
What happens if we skip:

1. Data collection: No data to learn from
2. Cleaning & Pre-processing --> garbage in, garbage out
3. Feature Engineering --> some algos will perform poor
4. Training --> bad performance on the test data.
5. Evaluation --> no clarity and awareness about the quality of the model
6. Deployment --> model becomes outdated and not useful.
```
