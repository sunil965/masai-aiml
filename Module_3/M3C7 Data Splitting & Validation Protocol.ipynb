#Train Test Split

```python
# 100 rows

# train -> 80 rows
# test -> 20 rows

# Training data --> teaching model the patterns
# test data --> evaluating the quality of the model.
```

```python
import pandas as pd

students = pd.DataFrame({
    "Hours" : [1,2,3,4,5,6,7,8],
    "Marks" : [20,30,40,50,60,70,80,90]
})

students
```

```python
x = students[["Hours"]]
y = students["Marks"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.2,
    random_state = 42
)
```

#Validation Set

```python
# Eg:

# train --> 70%
# validation -> 15%
# test -> 15%
```

```python
#purpose of each one of them:

1. Training data --> model learns the patterns
2. Validation data --> Hyperparameter tuning
3. Test data --> Evaluating the performance of the model.


Study --> mock test --> final exam

train --> validation --> test
```

```python
# train - 70%
# validation - 15%
# test - 15%
```

```python
# Idea:

# # 100 rows --> split(30%)a
# training rows = 70,
# temp rows = 30 --> split(50%) --> 15 validation, 15 test.
```

```python
from sklearn.model_selection import train_test_split

x_train, x_temp, y_train, y_temp = train_test_split(
    x,
    y,
    test_size = 0.3,
    random_state = 42)
```

```python
x_val, x_test, y_val, y_test = train_test_split(
    x_temp,
    y_temp,
    test_size = 0.5,
    random_state=  42
)
```
