```python
import pandas as pd

df = pd.DataFrame({
    "marks" :  [40, 60, 80, 100]
})
```

```python
df
```

#Min-Max Scaling (0 to 1) --> Normalization: transforming the data into a fixed range.

```python
from sklearn.preprocessing import MinMaxScaler

mn_mx_sc = MinMaxScaler()

df["Min_Max_Marks"] = mn_mx_sc.fit_transform(df[["marks"]])
df
```

#Standardization: Mean = 0, Std = 1

```python
from sklearn.preprocessing import StandardScaler

std_sc = StandardScaler()
df["std_sc_marks"] = std_sc.fit_transform(df[["marks"]])
df
```

#Which to use when?

```python
# Use Min-Max Scaler: (Normalization)

# 1. want the feature values to fall in a specific range( usually between 0 to 1)
# 2. When we need bounded inputs: essential for neural networks and for distance based algos like: KNN and K-means
# 3. when known meaningful limits exist --> for image processing: pixels ( 0 to 255), percentages ( 0 to 100), probabilityL: (0 to 1)
# 4. When the data is not normally distributed. --> min max scaler preserves the shape of the original distribution.
```

```python
# Use Standard Scaler (Standardization):
# 1. Works the best when the data is normally distributed (follows bell curve)
# 2. for Linear models and PCA --> Linear regression, logistic regression, SVM, PCA


# Imp note: When there is presence of outliers, both scalers might fail as they both are sensitive to outliers.
# eg: (1,2,3,4, 10000). In this case, when outliers are present, we prefer RobustScaler.
```

```python
df = pd.DataFrame({
    "Age" : [20,30,40],
    "Salary": [30000, 50000, 100000]
})

df
```

```python
sc = StandardScaler()

scaled = sc.fit_transform(df)
scaled_df = pd.DataFrame(scaled, columns = df.columns)

scaled_df
```

```python
df
```

```python
sc = MinMaxScaler()

scaled = sc.fit_transform(df)
mn_mx_df = pd.DataFrame(scaled, columns = df.columns)

mn_mx_df
```

```python
scaled_df
```

#Data Leakage

```python
# Never learn scaling parameters from the test data. Only use the training data to learn the scaler parameters.

# app1: data --> scaling --> train test split  --> data leakage
# app2: data --> train test split --> scaling --> correct
```

```python
df = pd.DataFrame({
    "Age":[20,25,30,35],
    "Salary":[30000,35000,90000,95000],
    "Approved":[0,0,1,1]
})

x = df[["Age", "Salary"]]
y = df["Approved"]

x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
                                                      x,
                                                      y,
                                                      test_size = 0.2,
                                                      random_state = 45)


sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```
