#Feature Scaling

```python
# Why scaling --> distance based algorithms might often give more importance to higher value featrues (salary in this case)

# Age     Salary
# 25      250000
# 28      350000

# 1. Min-Max Scaling
# 2. Standard Scaling
```

```python
import pandas as pd

df = pd.DataFrame({
    "Age":[20,30,40],
    "Salary":[30000,50000,100000]
})

df
```

#Min-Max Scaling

```python
# x --> (x - min)/(max - min) --> always ranges between 0 to 1
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

salary = pd.DataFrame({
    "Salary":[30000,50000,100000]
})
salary

```

```python
sc = MinMaxScaler()
df["Scaled_salary"] = sc.fit_transform(df[["Salary"]])
df
```

```python
(50000 - 30000)/(100000-30000)
```

#Standard Scaling

```python
# x --> (x - mean)/std --> Mean = 0, Std = 1
```

```python
from sklearn.preprocessing import StandardScaler

s_sc = StandardScaler()

df["std_scaled_salary"] = s_sc.fit_transform(df[["Salary"]])
df
```

#Creating more predictive features

```python
# eg: 1
# salary , years_of_exp  --> salary/years_of_exp

# orders, revenue --> reveue/order

# dob , curr-year--> age = curr_yr - dob

```

```python
df = pd.DataFrame({
    "Age":[18,25,35,50,65]
})

df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0,15,30,60,100],
    labels=["Teen","Young","Adult","Senior"]
)

df
```

#Date realted feature engineering

```python
df = pd.DataFrame({
    "Order_Date":[
        "2024-01-15",
        "2024-03-20"
    ]
})
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df
```

```python
df["year"] = df["Order_Date"].dt.year
df
```

```python
df["month"] = df["Order_Date"].dt.month
df
```

```python
df["day"] = df["Order_Date"].dt.day
df
```

```python
df["Weekday"] = df["Order_Date"].dt.day_of_week
df
```

```python
# is_weekend? 0-> monday

df["is_weekend"] = pd.cut(
    df["Weekday"],
    bins=[-1,5,7],
    labels=["No", "Yes"]
)

df
```

#Log Transformation

```python
# used for highly skewed data
```

```python
import numpy as np
salary = np.array([
    30000,
    40000,
    50000,
    1000000
])

# Log transformation reduces the effect of the extreme values

# log(10^3) = 3
# log(10^10) = 10


log_salary = np.log(salary)
log_salary
```

#Mini-Casestudy

```python
df = pd.DataFrame({
    "Age":[25,35,45],
    "Income":[50000,80000,100000],
    "Orders":[10,20,40]
})
```

```python
df
```

```python
# 1. income per order
df["income_per_order"] = df["Income"]/df["Orders"]
```

```python
df
```

```python
# 2. Above_avg --> if income is greater than the avg income of all the customers.

avg_income = df["Income"].mean()

df["above_avg"] = df["Income"] > avg_income
df
```

```python
# 3. create age group bucket.

df["age_group"] = pd.cut(df["Age"],
                         bins=[0,18,60,100],
                        labels=["Kid", "Adult", "Senior"]
                         )
df
```

```python
# 4. Scaled Salary --> Min-Max Scaling
df.describe()
```

```python
min_max_sc = MinMaxScaler()

df["min_max_income"] = min_max_sc.fit_transform(df[["Income"]])
df
```

```python
# 5 -> Standard scaling over income

std_sc = StandardScaler()

df["std_scaled_income"] = std_sc.fit_transform(df[["Income"]])
```
