# Combining DataFrames and Structural Joins

## Beginner-Friendly Google Colab Notebook

In real projects, useful information is often split across multiple datasets.

In this notebook, you will learn:

- Why datasets need to be combined
- How to combine similar tables using `pd.concat()`
- How to merge related tables using `pd.merge()`
- What inner, left, right, and outer joins mean
- How to check the final combined data

This notebook uses small examples so that the output is easy to understand.

## 1. Import Pandas

Pandas is the Python library we use for working with tabular data.

```python
import pandas as pd

print("Pandas imported successfully!")
```

## 2. Why Do We Combine DataFrames?

In companies, data is usually stored in different tables.

Example:

- One table may have student details
- Another table may have assessment scores

To analyze properly, we often need to combine these tables.

## 3. Create Two Simple Student DataFrames

```python
students = pd.DataFrame({
    "student_id": [101, 102, 103, 104],
    "student_name": ["Amit", "Priya", "John", "Sara"],
    "batch": ["A", "A", "B", "B"]
})

scores = pd.DataFrame({
    "student_id": [101, 102, 104, 105],
    "marks": [85, 78, 92, 66]
})

print("Students DataFrame:")
display(students)

print("Scores DataFrame:")
display(scores)
```

## 4. Understanding the Common Key

The common column is `student_id`.

This column helps Pandas match records from both DataFrames.

```python
print("Columns in students DataFrame:", students.columns.tolist())
print("Columns in scores DataFrame:", scores.columns.tolist())
```

# Part A: Combining Similar Tables using `concat()`

Use `concat()` when two DataFrames have the same or similar columns and you want to stack them together.

## 5. Example: Combining Monthly Sales Data

```python
jan_sales = pd.DataFrame({
    "order_id": [1, 2, 3],
    "product": ["Laptop", "Mouse", "Keyboard"],
    "amount": [55000, 700, 1500]
})

feb_sales = pd.DataFrame({
    "order_id": [4, 5, 6],
    "product": ["Monitor", "Mouse", "Laptop"],
    "amount": [12000, 800, 60000]
})

print("January Sales:")
display(jan_sales)

print("February Sales:")
display(feb_sales)
```

## 6. Row-wise Concatenation

This is like placing one Excel sheet below another Excel sheet.

```python
all_sales = pd.concat([jan_sales, feb_sales], ignore_index=True)

all_sales
```

## 7. Quick Check After Concatenation

```python
print("January rows:", jan_sales.shape[0])
print("February rows:", feb_sales.shape[0])
print("Combined rows:", all_sales.shape[0])
```

# Part B: Merging Related Tables using `merge()`

Use `merge()` when two DataFrames are related through a common key column.

## 8. Inner Join

An inner join keeps only the records that match in both DataFrames.

In this example, only students who are present in both `students` and `scores` will appear.

```python
inner_join = pd.merge(students, scores, on="student_id", how="inner")

inner_join
```

### Think About It

- Student `103` is in the students table but not in the scores table.
- Student `105` is in the scores table but not in the students table.
- Inner join removes unmatched records from both sides.

## 9. Left Join

A left join keeps all records from the left DataFrame.

Here, `students` is the left DataFrame, so all students will be kept.

```python
left_join = pd.merge(students, scores, on="student_id", how="left")

left_join
```

### What Does NaN Mean Here?

`NaN` means missing value.

Student `103` is present in the students table, but there is no score for this student in the scores table.

## 10. Right Join

A right join keeps all records from the right DataFrame.

Here, `scores` is the right DataFrame, so all score records will be kept.

```python
right_join = pd.merge(students, scores, on="student_id", how="right")

right_join
```

## 11. Outer Join

An outer join keeps all records from both DataFrames.

This is useful when you want to see all matched and unmatched records.

```python
outer_join = pd.merge(students, scores, on="student_id", how="outer")

outer_join
```

## 12. Summary of Join Types

| Join Type | Meaning | When to Use |
|---|---|---|
| Inner Join | Keeps only matching records | When only complete matched data is needed |
| Left Join | Keeps all records from left table | When left table is the main table |
| Right Join | Keeps all records from right table | When right table is the main table |
| Outer Join | Keeps all records from both tables | When you want to find all matches and gaps |

# Part C: Merging with Different Key Column Names

Sometimes the common key has different names in the two DataFrames.

```python
employee_details = pd.DataFrame({
    "emp_id": [1, 2, 3],
    "employee_name": ["Ravi", "Meena", "Kiran"]
})

salary_details = pd.DataFrame({
    "employee_id": [1, 2, 4],
    "salary": [50000, 62000, 45000]
})

print("Employee Details:")
display(employee_details)

print("Salary Details:")
display(salary_details)
```

## 13. Use `left_on` and `right_on`

Use this when the key column names are different.

```python
employee_salary = pd.merge(
    employee_details,
    salary_details,
    left_on="emp_id",
    right_on="employee_id",
    how="left"
)

employee_salary
```

# Part D: Handling Overlapping Column Names

Sometimes both DataFrames have columns with the same name. Pandas adds suffixes to avoid confusion.

```python
left_table = pd.DataFrame({
    "student_id": [101, 102],
    "name": ["Amit", "Priya"],
    "status": ["Active", "Active"]
})

right_table = pd.DataFrame({
    "student_id": [101, 102],
    "marks": [85, 78],
    "status": ["Submitted", "Pending"]
})

merged_status = pd.merge(
    left_table,
    right_table,
    on="student_id",
    how="inner",
    suffixes=("_student", "_assessment")
)

merged_status
```

# Part E: Validating Data After Combining

After combining data, always check whether the output looks correct.

## 14. Check Shape Before and After Merge

```python
print("Students shape:", students.shape)
print("Scores shape:", scores.shape)
print("Left join shape:", left_join.shape)
```

## 15. Check Missing Values

## 16. Find Students Without Scores

This is a common real-world use case.

```python
students_without_scores = left_join[left_join["marks"].isnull()]

students_without_scores
```

## 17. Check Duplicate Keys

Duplicate keys can create unexpected extra rows after merging.

```python
students["student_id"].duplicated().sum()
```

# Mini Project: Customer Orders Analysis

Now let us combine customer data with order data.

```python
customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "customer_name": ["Ananya", "Bharat", "Chitra", "Deepak"],
    "city": ["Bengaluru", "Hyderabad", "Pune", "Delhi"]
})

orders = pd.DataFrame({
    "order_id": [1001, 1002, 1003, 1004],
    "customer_id": [1, 2, 2, 5],
    "order_amount": [1200, 2500, 1800, 900]
})

print("Customers:")
display(customers)

print("Orders:")
display(orders)
```

## 18. Combine Customers and Orders

Question: Which customers have placed orders?

```python
customer_orders_inner = pd.merge(customers, orders, on="customer_id", how="inner")

customer_orders_inner
```

## 19. Keep All Customers and Add Orders Wherever Available

Question: Which customers have not placed any order?

```python
customer_orders_left = pd.merge(customers, orders, on="customer_id", how="left")

customer_orders_left
```

```python
customers_without_orders = customer_orders_left[customer_orders_left["order_id"].isnull()]

customers_without_orders
```

## 20. Keep All Customers and All Orders

Question: Are there any orders without customer details?

```python
customer_orders_outer = pd.merge(customers, orders, on="customer_id", how="outer")

customer_orders_outer
```

```python
orders_without_customer_details = customer_orders_outer[customer_orders_outer["customer_name"].isnull()]

orders_without_customer_details
```
