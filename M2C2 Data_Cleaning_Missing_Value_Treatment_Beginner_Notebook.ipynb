## 1. Import Pandas

Pandas helps us work with tabular data, similar to Excel sheets.

```python
import pandas as pd

print("Pandas imported successfully!")
```

## 2. Create a Small Messy Dataset

We will create a small employee dataset with common real-world problems:

- Missing values
- Salary stored as text
- Dates stored as text
- Extra spaces
- Inconsistent capitalization


```python
data = {
    "employee_id": [101, 102, 103, 104, 105, 106, 107],
    "name": ["Amit", "Priya", "John", "Kiran", "Sara", "Meena", "Ravi"],
    "department": ["IT", " HR", None, "Finance", "it", "Marketing", None],
    "age": [25, None, 30, 28, None, 32, 26],
    "salary": ["₹45,000", "₹52,000", None, "₹60,000", "₹48,000", "not available", "₹50,000"],
    "joining_date": ["2023-01-15", "2022/03/10", "15-04-2021", None, "2023-07-01", "Invalid Date", "2022-11-20"],
    "city": ["Bangalore", " hyderabad", "Mumbai", None, "bangalore", "Delhi ", "Chennai"]
}

df = pd.DataFrame(data)

df
```

## 3. Initial Data Inspection

Before cleaning, we first inspect the dataset.

Think of this like checking a room before cleaning it.

```python
# First five rows
df.head()
```

```python
# Number of rows and columns
df.shape
```

```python
# Basic information about columns and data types
df.info()
```

## 4. Identify Missing Values

Missing values usually appear as `NaN` or `None`.

In industry, this step is important because missing data can affect reports, dashboards, and ML models.

```python
# True means value is missing
df.isnull()
```

```python
# Count missing values in each column
df.isnull().sum()
```

```python
# Percentage of missing values in each column
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage
```

## 5. Understand Missing Value Treatment

There is no single perfect method for all missing values.

Simple beginner-friendly rules:

- If very few rows are missing, sometimes we can drop them.
- For numerical columns like age or salary, we can fill with mean or median.
- For text/category columns like department or city, we can fill with mode or `Unknown`.

For this session, we will use simple industry-friendly beginner methods.

## 6. Treat Missing Values in Text Columns

For columns like department and city, we will fill missing values with `Unknown`.

```python
df["department"] = df["department"].fillna("Unknown")
df["city"] = df["city"].fillna("Unknown")

df[["department", "city"]]
```

## 7. Treat Missing Values in Numeric Columns

The `age` column is numeric. We can fill missing age values using the median age.

Median is often preferred when we want a stable middle value.

```python
median_age = df["age"].median()
print("Median age:", median_age)

df["age"] = df["age"].fillna(median_age)

df[["name", "age"]]
```

## 8. Data Format Conversion: Salary Cleaning

The salary column currently contains values like `₹45,000`, so Pandas treats it as text.

Before calculations, we must convert it into numbers.

```python
# Check current salary values
df["salary"]
```

```python
# Remove currency symbol and comma, then convert to numeric
df["salary_cleaned"] = (
    df["salary"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["salary_cleaned"] = pd.to_numeric(df["salary_cleaned"], errors="coerce")

df[["salary", "salary_cleaned"]]
```

### Why `errors="coerce"`?

If a value cannot be converted into a number, Pandas converts it into `NaN`.

Example: `not available` cannot become a number, so it becomes missing.

```python
# Check missing values after salary conversion
df["salary_cleaned"].isnull().sum()
```

## 9. Fill Missing Salary Values

Now we can fill missing salary values using the median salary.

```python
median_salary = df["salary_cleaned"].median()
print("Median salary:", median_salary)

df["salary_cleaned"] = df["salary_cleaned"].fillna(median_salary)

df[["name", "salary", "salary_cleaned"]]
```

## 10. Data Format Conversion: Date Cleaning

Dates are often stored in different formats.

We will convert the `joining_date` column into a proper datetime format.

```python
df["joining_date_cleaned"] = pd.to_datetime(df["joining_date"], errors="coerce", dayfirst=True)

df[["joining_date", "joining_date_cleaned"]]
```

Invalid or missing dates become `NaT`, which means missing datetime value.

```python
df["joining_date_cleaned"].isnull().sum()
```

For this beginner session, we will fill missing dates with a placeholder date.

In real projects, the right treatment depends on business context.

```python
df["joining_date_cleaned"] = df["joining_date_cleaned"].fillna(pd.to_datetime("1900-01-01"))

df[["name", "joining_date_cleaned"]]
```

## 11. Basic Text Cleaning

Text columns may have:

- Extra spaces
- Different capitalization
- Same value written in different ways

Example: `IT`, `it`, and ` IT` should be treated consistently.

```python
df["department_cleaned"] = df["department"].astype(str).str.strip().str.title()
df["city_cleaned"] = df["city"].astype(str).str.strip().str.title()

df[["department", "department_cleaned", "city", "city_cleaned"]]
```

## 12. Convert Age to Integer

After filling missing values, we can convert age from float to integer.

```python
df["age"] = df["age"].astype(int)

df[["name", "age"]]
```

## 13. Create Final Cleaned Dataset

Now we will keep only useful cleaned columns.

```python
cleaned_df = df[[
    "employee_id",
    "name",
    "department_cleaned",
    "age",
    "salary_cleaned",
    "joining_date_cleaned",
    "city_cleaned"
]]

cleaned_df = cleaned_df.rename(columns={
    "department_cleaned": "department",
    "salary_cleaned": "salary",
    "joining_date_cleaned": "joining_date",
    "city_cleaned": "city"
})

cleaned_df
```

## 14. Validate the Cleaned Dataset

After cleaning, always validate.

Basic validation checklist:

- Are missing values handled?
- Are data types correct?
- Does the dataset look clean?
- Can we perform calculations?

```python
# Check missing values again
cleaned_df.isnull().sum()
```

```python
# Check final data types
cleaned_df.info()
```

```python
# Can we calculate average salary now?
average_salary = cleaned_df["salary"].mean()
print("Average Salary:", average_salary)
```

```python
# Can we calculate average age now?
average_age = cleaned_df["age"].mean()
print("Average Age:", average_age)
```

## 15. Mini Practice Task

Try this with the dataset below.

Tasks:

1. Create the DataFrame
2. Find missing values
3. Fill missing product category with `Unknown`
4. Convert price to numeric
5. Fill missing price with median price
6. Convert order date to datetime
7. Validate the cleaned dataset

```python
practice_data = {
    "order_id": [1, 2, 3, 4, 5],
    "product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headphones"],
    "category": ["Electronics", None, "Accessories", " Electronics ", None],
    "price": ["₹55,000", "₹700", None, "₹12,000", "not available"],
    "order_date": ["2024-01-10", "11/02/2024", None, "Invalid Date", "2024-03-15"]
}

practice_df = pd.DataFrame(practice_data)
practice_df
```

### Practice Solution

```python
# 1. Check missing values
practice_df.isnull().sum()
```

```python
# 2. Clean category
practice_df["category"] = practice_df["category"].fillna("Unknown")
practice_df["category"] = practice_df["category"].astype(str).str.strip().str.title()

# 3. Clean price and convert to numeric
practice_df["price_cleaned"] = (
    practice_df["price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)
practice_df["price_cleaned"] = pd.to_numeric(practice_df["price_cleaned"], errors="coerce")

# 4. Fill missing price with median
practice_df["price_cleaned"] = practice_df["price_cleaned"].fillna(practice_df["price_cleaned"].median())

# 5. Convert order_date to datetime
practice_df["order_date_cleaned"] = pd.to_datetime(practice_df["order_date"], errors="coerce", dayfirst=True)

practice_df
```

```python
# Final validation
practice_df.info()
print("\nMissing values after cleaning:")
print(practice_df.isnull().sum())
```
