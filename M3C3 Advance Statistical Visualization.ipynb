#Correlationship Score

```python
# Correlation score: ranges between -1 to +1

# +1 --> Strongly positive correlated
# -1 --> strongly negative correlated
# 0 --> no correlation
```

```python
corr_score = tips.corr()
```

```python
corr_score = tips.corr(numeric_only = True)
```

```python
corr_score
```

```python
sns.heatmap(corr_score)
```

```python
sns.heatmap(corr_score,
            annot = True)
```

```python
sns.heatmap(corr_score,
            annot = True,
            cmap = "Blues")
```

```python
sns.heatmap(corr_score,
            annot = True,
            cmap = "Greens")
```

```python
sns.heatmap(corr_score,
            annot = True,
            cmap = "Reds")
```
