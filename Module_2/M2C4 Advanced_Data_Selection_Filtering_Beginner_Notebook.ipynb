# Advanced Data Selection and Filtering

## Beginner-Friendly Hands-on Notebook

Topics covered:

- Selecting rows and columns
- Applying filters
- Using logical conditions
- Sorting data
- Reordering information


## Step 1: Import Pandas

```python

import pandas as pd

print("Pandas imported successfully!")
```

## Step 2: Create Dataset

```python

data = {
    "Employee": ["Amit", "Priya", "John", "Kiran", "Sara", "Ravi"],
    "Department": ["HR", "IT", "Finance", "IT", "Marketing", "HR"],
    "Salary": [45000, 70000, 65000, 72000, 50000, 48000],
    "Experience": [2, 5, 4, 6, 3, 1]
}

df = pd.DataFrame(data)

df
```

## Step 3: Selecting Columns

```python

df["Employee"]
```

```python

df[["Employee", "Salary"]]
```

## Step 4: Selecting Rows

```python

df.loc[0]
```

```python

df.loc[0:2]
```

## Step 5: Filtering Data

```python

df[df["Salary"] > 60000]
```

```python

df[df["Department"] == "IT"]
```

## Step 6: Multiple Conditions

```python

df[(df["Salary"] > 50000) & (df["Experience"] > 3)]
```

```python

df[(df["Department"] == "HR") | (df["Department"] == "Marketing")]
```

## Step 7: Sorting Data

```python

df.sort_values(by="Salary")
```

```python

df.sort_values(by="Salary", ascending=False)
```

```python

df.sort_values(by=["Department", "Salary"])
```

## Step 8: Reordering Columns

```python

df[["Department", "Employee", "Salary", "Experience"]]
```
