# Data Aggregation and Grouping with Pandas

## Beginner-Friendly Google Colab Notebook

In this notebook, we will learn how to summarize data using Pandas `groupby()`.

- How to calculate category-wise count, sum, mean, min, and max

## 1. Import Pandas

Pandas is the main Python library used for working with structured/tabular data.


```python

import pandas as pd

print("Pandas imported successfully!")
```


## 2. Create a Simple Dataset

We will use a small employee dataset.

Each row represents one employee.
Each column gives information about that employee.


```python

employee_data = {
    "Employee": ["Amit", "Priya", "John", "Kiran", "Sara", "Ravi", "Meena", "Arjun", "Fatima", "Neha"],
    "Department": ["HR", "IT", "Finance", "IT", "Marketing", "HR", "Finance", "IT", "Marketing", "Finance"],
    "City": ["Bengaluru", "Hyderabad", "Bengaluru", "Pune", "Hyderabad", "Pune", "Bengaluru", "Hyderabad", "Pune", "Bengaluru"],
    "Gender": ["Male", "Female", "Male", "Male", "Female", "Male", "Female", "Male", "Female", "Female"],
    "Salary": [45000, 70000, 65000, 72000, 50000, 48000, 62000, 76000, 54000, 68000],
    "Experience": [2, 5, 4, 6, 3, 1, 4, 7, 3, 5]
}

df = pd.DataFrame(employee_data)

df
```


## 3. Quick Data Inspection

Before grouping, we should understand the dataset.


```python

df.head()
```

```python

print("Rows and Columns:", df.shape)
print("Column Names:", df.columns.tolist())
```

```python

df.info()
```


## 4. What is Aggregation?

Aggregation means summarizing many rows into fewer meaningful results.

### Example
Instead of checking every employee salary one by one, we may ask:

**What is the average salary in each department?**

This is common in industry reporting and dashboarding.



## 5. Basic `groupby()` Example

Let us calculate the average salary by department.

Mental model:

1. Split data by Department
2. Pick Salary column
3. Calculate average salary for each group


```python

df.groupby("Department")["Salary"].mean()
```


### Reading the Output

The output tells us the average salary for each department.

This is more useful than looking at all individual employee records.



## 6. Count Records in Each Group

`count()` helps us know how many records are present in each category.

Example question:

**How many employees are there in each department?**


```python

df.groupby("Department")["Employee"].count()
```


## 7. Total Value Using `sum()`

`sum()` helps us calculate total values.

Example question:

**What is the total salary cost for each department?**


```python

df.groupby("Department")["Salary"].sum()
```


## 8. Minimum and Maximum Values

`min()` gives the lowest value.
`max()` gives the highest value.

Example question:

**What is the lowest and highest salary in each department?**


```python

df.groupby("Department")["Salary"].min()
```

```python

df.groupby("Department")["Salary"].max()
```


## 9. Multiple Aggregations with `.agg()`

In real analysis, we often need multiple summaries together.

Example question:

**For each department, show employee count, average salary, minimum salary, and maximum salary.**


```python

df.groupby("Department")["Salary"].agg(["count", "mean", "min", "max"])
```


## 10. Aggregating Multiple Columns

We can summarize more than one numerical column.

Example:
- Average salary by department
- Average experience by department


```python

df.groupby("Department")[["Salary", "Experience"]].mean()
```


## 11. Grouping by City

Grouping is not limited to departments.

Example question:

**What is the average salary in each city?**


```python

df.groupby("City")["Salary"].mean()
```


## 12. Grouping by Two Columns

Sometimes one category is not enough.

Example question:

**What is the average salary by Department and Gender?**


```python

df.groupby(["Department", "Gender"])["Salary"].mean()
```


## 13. Using `reset_index()`

After grouping, Pandas may show the grouping column as an index.

`reset_index()` converts the grouped result back into a normal DataFrame.

This makes the output easier to read, sort, export, or plot later.


```python

department_summary = df.groupby("Department")["Salary"].mean().reset_index()

department_summary
```


## 14. Sorting Grouped Results

After summarizing, we often sort the result to find top or bottom categories.

Example question:

**Which department has the highest average salary?**


```python

department_summary.sort_values(by="Salary", ascending=False)
```


## 15. Rename Aggregated Columns

Renaming columns makes summaries easier to understand.


```python

department_summary = department_summary.rename(columns={"Salary": "Average_Salary"})

department_summary
```


# 16. Industry-Style Example: Sales Dataset

Now let us use a simple sales dataset.

This is closer to real business analysis.


```python

sales_data = {
    "Order_ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "City": ["Bengaluru", "Hyderabad", "Bengaluru", "Pune", "Hyderabad", "Pune", "Bengaluru", "Hyderabad", "Pune", "Bengaluru"],
    "Product_Category": ["Electronics", "Clothing", "Electronics", "Grocery", "Electronics", "Clothing", "Grocery", "Grocery", "Electronics", "Clothing"],
    "Quantity": [2, 5, 1, 10, 3, 4, 8, 6, 2, 7],
    "Sales_Amount": [50000, 15000, 30000, 8000, 60000, 12000, 9000, 7000, 40000, 21000]
}

sales_df = pd.DataFrame(sales_data)

sales_df
```


## 17. Total Sales by City

Business question:

**Which city generated the highest total sales?**


```python

city_sales = sales_df.groupby("City")["Sales_Amount"].sum().reset_index()

city_sales
```

```python

city_sales.sort_values(by="Sales_Amount", ascending=False)
```


## 18. Total Sales by Product Category

Business question:

**Which product category generated the highest total sales?**


```python

category_sales = sales_df.groupby("Product_Category")["Sales_Amount"].sum().reset_index()

category_sales.sort_values(by="Sales_Amount", ascending=False)
```


## 19. Multiple Aggregations on Sales

Business question:

**For each product category, find total sales, average sales, and number of orders.**


```python

sales_df.groupby("Product_Category")["Sales_Amount"].agg(["count", "sum", "mean", "min", "max"])
```


## 20. Grouping Sales by City and Product Category

Business question:

**How much sales came from each product category in each city?**


```python

sales_df.groupby(["City", "Product_Category"])["Sales_Amount"].sum()
```


## 21. Cleaner Multi-Column Grouping Output

Let us use `reset_index()` to make the previous output easier to read.


```python

city_category_sales = sales_df.groupby(["City", "Product_Category"])["Sales_Amount"].sum().reset_index()

city_category_sales
```


## 22. Sort Multi-Column Grouping Output

Now we can sort to find the highest city-category combinations.


```python

city_category_sales.sort_values(by="Sales_Amount", ascending=False)
```
