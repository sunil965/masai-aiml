## 1. Why NumPy? Numerical Computing Mindset

In real-world data work, we often need to perform the same operation on many numbers:

- Add bonus marks to all students
- Apply GST/tax to all product prices
- Convert temperatures for many sensor readings
- Calculate revenue from price and quantity

Plain Python lists are useful, but NumPy makes numerical operations faster and cleaner.


```python
# Import NumPy using the standard industry convention
import numpy as np

print('NumPy imported successfully!')
print('NumPy version:', np.__version__)
```

```python
# Python list approach: add 5 bonus marks to each student
marks_list = [45, 67, 82, 39, 90]

bonus_marks_list = []
for mark in marks_list:
    bonus_marks_list.append(mark + 5)

print('Original marks using Python list:', marks_list)
print('After adding bonus using loop:', bonus_marks_list)
```

```python
# NumPy vectorized approach: add 5 bonus marks to each student
marks_array = np.array([45, 67, 82, 39, 90])
bonus_marks_array = marks_array + 5

print('Original marks using NumPy array:', marks_array)
print('After adding bonus using NumPy:', bonus_marks_array)
```

> **Key idea:** Vectorization means applying an operation to many values at once without writing a manual loop.

## 4. Selecting Data: Indexing, Slicing, and Filtering

In real-world work, we rarely use all data at once. We often need to select specific values, rows, columns, or values that match a condition.


### 4.1 Indexing in 1D Arrays

Index positions start from 0.


```python
marks = np.array([72, 85, 64, 90, 58])

print('All marks:', marks)
print('First student mark:', marks[0])
print('Third student mark:', marks[2])
print('Last student mark:', marks[-1])
```

### 4.2 Slicing in 1D Arrays

Slicing selects a range of values.


```python
print('First three marks:', marks[0:3])
print('Marks from index 2 onwards:', marks[2:])
print('Marks except last value:', marks[:-1])
```

### 4.3 Indexing in 2D Arrays

For 2D arrays, use `[row, column]`.


```python
print(score_table)

print('\nFirst student, Python score:', score_table[0, 0])
print('Fourth student, Communication score:', score_table[3, 2])
```

### 4.4 Selecting Rows and Columns


```python
# Select complete row: all subject scores of first student
print('First student all scores:', score_table[0, :])

# Select complete column: Python scores of all students
print('Python scores of all students:', score_table[:, 0])

# Select Math scores of all students
print('Math scores of all students:', score_table[:, 1])
```

### 4.5 Boolean Filtering

Boolean filtering helps us select values that satisfy a condition.


```python
marks = np.array([72, 85, 64, 90, 58, 33, 49])

pass_mask = marks >= 50
print('Marks:', marks)
print('Pass mask:', pass_mask)
print('Passing marks only:', marks[pass_mask])
```

```python
# Same filtering in one line
passing_marks = marks[marks >= 50]
failed_marks = marks[marks < 50]

print('Passing marks:', passing_marks)
print('Failed marks:', failed_marks)
```

## 5. Vectorized Operations: Replacing Loops with Array Operations

Vectorization is one of the most important NumPy concepts.

It helps us write shorter and cleaner code for numerical operations.


### 5.1 Array-to-Scalar Operations

Apply one number to every value in the array.


```python
marks = np.array([72, 85, 64, 90, 58])

print('Original marks:', marks)
print('After adding 5 bonus marks:', marks + 5)
print('After doubling marks:', marks * 2)
print('Marks as percentage of 100:', marks / 100)
```

### 5.2 Array-to-Array Operations

When two arrays have matching positions, NumPy can operate element by element.


```python
prices = np.array([100, 250, 75, 300, 150])
quantities = np.array([2, 1, 5, 1, 3])

revenue = prices * quantities

print('Prices:', prices)
print('Quantities:', quantities)
print('Revenue per product:', revenue)
```

### 5.3 Business Rule Example: Apply GST/Tax


```python
prices = np.array([100, 250, 75, 300, 150])
gst_rate = 0.18

final_prices = prices + (prices * gst_rate)

print('Original prices:', prices)
print('Final prices after 18% GST:', final_prices)
```

### 5.4 Using `np.where` for Simple Conditions

`np.where(condition, value_if_true, value_if_false)` is useful for simple classification or transformation.


```python
marks = np.array([72, 85, 64, 90, 58, 33, 49])

result = np.where(marks >= 50, 'Pass', 'Fail')

print('Marks:', marks)
print('Result:', result)
```

```python
# Add 5 bonus marks, but cap the maximum score at 100
marks = np.array([96, 82, 77, 99, 65])
bonus_marks = marks + 5
capped_marks = np.where(bonus_marks > 100, 100, bonus_marks)

print('Original marks:', marks)
print('After bonus:', bonus_marks)
print('After capping at 100:', capped_marks)
```

## 6. Useful Mathematical and Statistical Functions

NumPy provides ready-made functions to summarize numerical data.


```python
marks = np.array([72, 85, 64, 90, 58, 33, 49])

print('Marks:', marks)
print('Total:', np.sum(marks))
print('Average:', np.mean(marks))
print('Minimum:', np.min(marks))
print('Maximum:', np.max(marks))
print('Median:', np.median(marks))
print('Standard deviation:', np.std(marks))
print('Rounded standard deviation:', np.round(np.std(marks), 2))
```

### Axis-Based Operations

For 2D arrays:

- `axis=0` means column-wise calculation
- `axis=1` means row-wise calculation


```python
# Rows: students
# Columns: Python, Math, Communication
score_table = np.array([
    [72, 80, 75],
    [85, 78, 82],
    [64, 70, 68],
    [90, 88, 92],
    [58, 60, 62]
])

print(score_table)
```

```python
student_averages = np.mean(score_table, axis=1)
subject_averages = np.mean(score_table, axis=0)

print('Average score per student:', student_averages)
print('Average score per subject:', subject_averages)
```

```python
subject_names = np.array(['Python', 'Math', 'Communication'])

for subject, avg in zip(subject_names, subject_averages):
    print(subject, 'average =', round(avg, 2))
```

## 7. Reshaping Data for Analysis and ML Readiness

Sometimes raw data comes as a single sequence, but we need to arrange it as rows and columns.

Important rule: The total number of values must remain the same.


```python
monthly_sales = np.array([120, 135, 150, 160, 170, 165, 180, 190, 200, 210, 220, 230])

print('Monthly sales as 1D array:')
print(monthly_sales)
print('Shape:', monthly_sales.shape)
```

```python
# Reshape 12 monthly values into 4 quarters x 3 months
quarterly_sales = monthly_sales.reshape(4, 3)

print('Quarterly sales table:')
print(quarterly_sales)
print('Shape:', quarterly_sales.shape)
```

```python
# Flatten a 2D array back to 1D
flattened_sales = quarterly_sales.flatten()

print('Flattened sales:')
print(flattened_sales)
print('Shape:', flattened_sales.shape)
```

### Common Reshape Error

The following example is intentionally shown with error handling.

We cannot reshape 12 values into 5 rows and 3 columns because 5 × 3 = 15 values are required.


```python
try:
    wrong_shape = monthly_sales.reshape(5, 3)
except ValueError as error:
    print('Error message:', error)
    print('Reason: 12 values cannot fit into a 5 x 3 table.')
```

## 8. Mini Industry-Style Practice: Product Sales Analysis

In this mini case study, we will combine the must-have NumPy concepts:

- Array creation
- Inspection
- Vectorized revenue calculation
- Filtering
- Aggregation
- Plain-English interpretation


```python
# Product sales data
product_names = np.array(['Notebook', 'Pen', 'Bag', 'Bottle', 'Calculator'])
prices = np.array([80, 10, 600, 250, 450])
quantities_sold = np.array([120, 500, 35, 80, 45])

print('Products:', product_names)
print('Prices:', prices)
print('Quantities sold:', quantities_sold)
```

```python
# Step 1: Inspect arrays
print('Product names shape:', product_names.shape)
print('Prices shape:', prices.shape)
print('Quantities shape:', quantities_sold.shape)
```

```python
# Step 2: Calculate revenue using vectorization
revenue = prices * quantities_sold

print('Revenue per product:', revenue)
```

```python
# Step 3: Show revenue with product names
for product, rev in zip(product_names, revenue):
    print(product, 'revenue = Rs.', rev)
```

```python
# Step 4: Calculate total and average revenue
total_revenue = np.sum(revenue)
average_revenue = np.mean(revenue)

print('Total revenue = Rs.', total_revenue)
print('Average revenue = Rs.', round(average_revenue, 2))
```

```python
# Step 5: Find products with revenue above average
high_revenue_mask = revenue > average_revenue
high_revenue_products = product_names[high_revenue_mask]
high_revenue_values = revenue[high_revenue_mask]

print('Products with above-average revenue:', high_revenue_products)
print('Their revenues:', high_revenue_values)
```

```python
# Step 6: Find the top revenue product
top_index = np.argmax(revenue)

print('Top revenue product:', product_names[top_index])
print('Top revenue value: Rs.', revenue[top_index])
```

### Plain-English Interpretation

The code above helps answer business questions such as:

- How much revenue did each product generate?
- What is the total revenue?
- Which products performed above average?
- Which product generated the highest revenue?


## Final Guided Practice

Ask learners to complete this section independently first. Then reveal/discuss the solution.

### Task
You are given marks of students in three subjects: Python, Math, and Communication.

Find:

1. Average score of each student
2. Average score of each subject
3. Students who passed overall, assuming average >= 50
4. Highest average score


```python
# Final Practice Data
student_names = np.array(['Asha', 'Rahul', 'Meena', 'Kiran', 'Fatima'])
scores = np.array([
    [72, 80, 75],
    [45, 50, 48],
    [88, 92, 85],
    [60, 55, 58],
    [35, 42, 40]
])

print('Student names:', student_names)
print('Scores:')
print(scores)
```

```python
# Solution
student_avg = np.mean(scores, axis=1)
subject_avg = np.mean(scores, axis=0)
passed_students = student_names[student_avg >= 50]
highest_avg = np.max(student_avg)
topper_name = student_names[np.argmax(student_avg)]

print('Average score of each student:')
for name, avg in zip(student_names, student_avg):
    print(name, '=', round(avg, 2))

print('\nAverage score of each subject:')
for subject, avg in zip(['Python', 'Math', 'Communication'], subject_avg):
    print(subject, '=', round(avg, 2))

print('\nStudents who passed overall:', passed_students)
print('Highest average score:', round(highest_avg, 2))
print('Top student:', topper_name)
```
