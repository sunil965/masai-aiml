## Step 1: Import Pandas

```python

import pandas as pd

print("Pandas imported successfully!")
```


## Step 2: Creating a Simple DataFrame


```python

data = {
    "Name": ["Harsha", "Rahul", "Sneha", "Aman"],
    "Age": [24, 25, 23, 26],
    "Marks": [88, 76, 91, 67]
}

df = pd.DataFrame(data)

df
```


## Step 3: Understanding Rows and Columns


```python

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```


## Step 4: Creating and Loading External Dataset


```python

sample_data = pd.DataFrame({
    "Employee": ["Amit", "Priya", "John", "Kiran", "Sara"],
    "Department": ["HR", "IT", "Finance", "IT", "Marketing"],
    "Salary": [45000, 70000, 65000, 72000, 50000]
})

sample_data.to_csv("employees.csv", index=False)

print("CSV File Created")
```

```python

employees_df = pd.read_csv("employees.csv")

employees_df
```


## Step 8: Filtering Data


```python

employees_df[employees_df["Salary"] > 60000]
```

```python

employees_df[employees_df["Department"] == "IT"]
```


## Step 9: Missing Values Awareness
